```python
df1 = df1.drop('Region', axis = 'columns')
df1 = df1.drop('Happiness Rank', axis = 'columns')
df1 = df1.drop('Standard Error', axis = 'columns')
df1 = df1.drop('Dystopia Residual', axis = 'columns')
```

```python
df2 = df2.drop('Region', axis = 'columns')
df2 = df2.drop('Happiness Rank', axis = 'columns')
df2 = df2.drop('Lower Confidence Interval', axis = 'columns')
df2 = df2.drop('Upper Confidence Interval', axis = 'columns')
df2 = df2.drop('Dystopia Residual', axis = 'columns')
```

```python
df3 = df3.drop('Happiness.Rank', axis = 'columns')
df3 = df3.drop('Whisker.high', axis = 'columns')
df3 = df3.drop('Whisker.low', axis = 'columns')
df3 = df3.drop('Dystopia.Residual', axis = 'columns')
```

```python
df4 = df4.drop('Overall rank', axis = 'columns')
```

```python
df5 = df5.drop('Overall rank', axis = 'columns')
```

```python
def swap_df_columns(df, col1, col2):
    col_list = list(df)

    a, b = col_list.index(col1), col_list.index(col2)

    col_list[b], col_list[a] = col_list[a], col_list[b]

    df = df[col_list]

    return df
```

```python
df3 = swap_df_columns(df3, 'Generosity', 'Trust..Government.Corruption.')
df4 = swap_df_columns(df4, 'Generosity', 'Perceptions of corruption')
df5 = swap_df_columns(df5, 'Generosity', 'Perceptions of corruption')
```

```python
def renombrar_columnas(df, nombres_columnas):
    df.columns = nombres_columnas
    return df
```

```python
nombres_columnas_df1 = df1.columns.tolist()

df2 = renombrar_columnas(df2, nombres_columnas_df1)
df3 = renombrar_columnas(df3, nombres_columnas_df1)
df4 = renombrar_columnas(df4, nombres_columnas_df1)
df5 = renombrar_columnas(df5, nombres_columnas_df1)
```

```python
def add_year(df, year):
    df['Year'] = year
    return df
```

```python
df1 = add_year(df1, 2015)
df2 = add_year(df2, 2016)
df3 = add_year(df3, 2017)
df4 = add_year(df4, 2018)
df5 = add_year(df5, 2019)
```

```python
happiness = pd.concat([df1, df2, df3, df4, df5], axis=0)
```

```python
happiness
```

```python
columnas_numericas = happiness.select_dtypes(include=np.number).columns

num_columnas = len(columnas_numericas)
num_filas = int(np.ceil(num_columnas / 2))

fig, axs = plt.subplots(num_filas, 2, figsize=(12, num_filas * 5))

for i, columna in enumerate(columnas_numericas):
    fila = i // 2
    columna_sub = i % 2
    ax = axs[fila, columna_sub]

    datos = happiness[columna]
    res = stats.probplot(datos, dist="norm", plot=ax)
    ax.get_lines()[1].set_color('red')  # Línea de referencia para la distribución normal
    ax.set_title(f'QQ Plot - {columna}')
    ax.set_xlabel('Cuantiles teóricos')
    ax.set_ylabel('Cuantiles de los datos')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,  7))
happiness.boxplot(column=['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity'])
plt.show()
```

```python
columnas_numericas = happiness.select_dtypes(include=np.number).columns

num_columnas = len(columnas_numericas)
num_filas = int(np.ceil(num_columnas / 2))

fig, axs = plt.subplots(num_filas, 2, figsize=(12, num_filas * 5))

for i, columna in enumerate(columnas_numericas):
    fila = i // 2
    columna_sub = i % 2
    ax = axs[fila, columna_sub]

    sns.kdeplot(data=happiness[columna], ax=ax, fill=True, color='green')
    ax.set_title(f'{columna}')
    ax.set_xlabel('Valor')
    ax.set_ylabel('Densidad')

plt.tight_layout()
plt.show()
```

```python
average_happiness = happiness.groupby('Country')['Happiness Score'].mean()

average_happiness = average_happiness.sort_values(ascending=False)

top_5_happiest_countries = average_happiness.head(5)
top_5_unhappiest_countries = average_happiness.tail(5)

selected_countries = top_5_happiest_countries.index.union(top_5_unhappiest_countries.index)
filtered_data = happiness[happiness['Country'].isin(selected_countries)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_data, x='Year', y='Happiness Score', hue='Country', marker='o')
plt.title('Puntaje de Felicidad de los 5 Países Más Felices y 5 Países Más Infelices (2015-2019)')
plt.xlabel('Año')
plt.ylabel('Puntaje de Felicidad')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
